```python
import requests
from secrets_config import api_key
import pandas as pd
```

```python
# Variables: Note: these are case-sensitive
symbol = 'BTC'
# could also do: symbol = 'ETH'
market = 'USD'
```

```python
def extract_crypto_rates(
    symbol:str,
    market:str,
    api_key:str
)->pd.DataFrame:
    base_url = f'https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol={symbol}&market={market}&apikey={api_key}'
    r = requests.get(base_url)
    response_data = r.json()
    df = pd.DataFrame(response_data['Time Series (Digital Currency Daily)']).transpose().reset_index()
    df = df.rename(columns={ df.columns[0]: "Date" })
    df['Symbol'] = f'{symbol}'
    df['Market'] = f'{market}'
    return df
```

```python
extract_crypto_rates(symbol=symbol, market=market, api_key=api_key)
```

```python
def extract_fed_funds_rate(
    api_key:str
)->pd.DataFrame:
    base_url = f'https://www.alphavantage.co/query?function=FEDERAL_FUNDS_RATE&interval=daily&apikey={api_key}' 
    r = requests.get(base_url)
    response_data = r.json()
    df = pd.DataFrame(response_data['data'])
    df = df.rename(columns={ df.columns[0]: "Date" })
    df['Symbol'] = f'{symbol}'
    df['Market'] = f'{market}'
    return df
```

```python
extract_fed_funds_rate(api_key=api_key)
```
